Не трудно заметить, что тип данных для общего столбца `puNumber` разный. Посмотрим на первые пять строчек датасета и идетифицируем необходимый тип данных

```python
df_mkrf_movies.head()
```

Колонка `puNumber` это число, а точнее номер прокатного удостоверения, предположим, содержащее исключительно цифры. Остается открытым вопрос о том, как были заполнены значения для тех фильмов у которых нет такого номера. Примем решение о преобразовании в целочисленный тип данных `int`

```python
# df_mkrf_movies['puNumber'] = df_mkrf_movies['puNumber'].astype(int)
```

Что и следовало ожидать, сообщение об ошибке `ValueError: invalid literal for int() with base 10: 'нет'`, говорит нам о том, что в колонке присутствуют и символьные и числовые данные. В данном случае, просмотреть уникальные данные будет не лучшей идеей, так как каждый номер априори уникален. Напишем функцию для обработки данной ошибки и выполним преобразование.

```python
df_mkrf_movies = df_mkrf_movies.apply(lambda x: x.replace({'нет':'0'}, regex=True))
```

```python
df_mkrf_movies['puNumber'] = df_mkrf_movies['puNumber'].apply(int)
```

```python
df_mkrf_movies.info()
```

Теперь датасеты можно объединять

```python
df = df_mkrf_movies.merge(df_mkrf_shows, how='left')
```

```python
df.info()
```

### Типы данных

- Проверьте типы данных в датафрейме и преобразуйте их там, где это необходимо.

Преобразуем `show_start_date` из строки в формат даты и времени.

```python
df['show_start_date'] = pd.to_datetime(df['show_start_date'], format='%Y-%m-%dT%H:%M:%S.000Z')
```

Для дальнейшего анализа может потребоваться фозрастная аудитория фильма, сейчас данные выглядят не компьютор-френдли, с такими данными трудно работать. Выполним замену данных и преобразуем их в `int`

```python
df['age_restriction'].unique()
```

```python
df = df.apply(lambda x: x.replace({'«18+» - запрещено для детей':'18', '«6+» - для детей старше 6 лет':'6',
                                  '«12+» - для детей старше 12 лет':'12', '«16+» - для детей старше 16 лет':'16',
                                  '«0+» - для любой зрительской аудитории':'0'}))
```

```python
df['age_restriction'] = df['age_restriction'].apply(int)
```

Осталось привести занчения рейтинга к вещественному типу данных

```python
df['ratings'].unique()
```

Из набора уникальных значений видно, что есть пропуски в данных, посмотрим сколько их

```python
df['ratings'].isna().sum()
```

Пропусков достаточно много, но удалять или заменять их нельзя, чтобы не искревлять статистику.

```python
mask = df['ratings'].str.endswith('%', na=False)
df.loc[mask,'ratings'] = [(i[0]+'.'+i[1]) for i in df.loc[mask,'ratings']]
```

```python
df['ratings'].unique()
```

```python
df['ratings'] = df['ratings'].apply(float)
```

```python
df.info()
```

На этом этап преобразования данных можно считать завершенным.

```python
df['film_studio'].unique()
```

Уберем потенциальные пробелы в начале и в конце названий студий

```python
end_space = df['film_studio'].str.endswith(' ', na=False)
start_space = df['film_studio'].str.startswith(' ', na=False)
df.loc[end_space,'film_studio'] = [i[:-1] for i in df.loc[end_space,'film_studio']]
df.loc[start_space,'film_studio'] = [i[1:] for i in df.loc[start_space,'film_studio']]
```

Есть предположение, что каждый из фильмов имеет разную запись студии. Проверим явные дубликаты.

```python
df['film_studio'].duplicated().sum()
```

Лишь порядка 26% похожих записей. Исправлять каждую это потеря огромного количества времни, но можно обратить внимание на сериалы и фильмы имеющие сиквелы. Попробуем выполнить поиск с условием, что после каждого названия сериала или фильма есть `:` после которого идёт название серии или части и найдём названия студий.

```python
tmask = list(dict.fromkeys([i[:i.find(':')] for i in df['title'].unique() if ':' in i]))
studio_key = {}
for i in tmask:
    studio_key[i] = df.loc[df['title'].str.startswith(i, na=False),'film_studio']
studio_key
```

С получившимся набором данных трудно работать, так как не известа история владения фильмов той или иной студии. На первый взгляд большинство студий записаны верно. Более детальный анализ предлагается приостановить.

```python
def age_box_office(year):
    f = {'age': [df['age_restriction'][i] for i in range(len(df)) if df['rental_year'][i] == year],
        'box_office': [df['box_office'][i] for i in range(len(df)) if df['rental_year'][i] == year]}
    return pd.DataFrame(data=f).sort_values(by='age').reset_index()
```

```python
def year_box_office(age):
    f = {'year': [df['rental_year'][i] for i in range(len(df)) if df['age_restriction'][i] == age],
        'box_office': [df['box_office'][i] for i in range(len(df)) if df['age_restriction'][i] == age]}
    return pd.DataFrame(data=f).sort_values(by='year').reset_index()
```
